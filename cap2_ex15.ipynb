### Exercício 15. Comprimento de Parafusos de 3/4 polegada
#### Use a distribuição de frequência do Exercício 23, na seção 2-2, para contruir o histograma.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# versão
print(f"Versão do Pandas: {pd.__version__}")
print(f"Versão do Matplotlib: {matplotlib.__version__}")
```

```python
# valores do exercício
valores = [0.757, 0.723, 0.754, 0.737, 0.757, 0.741, 0.722, 0.741, 0.743, 0.742, 0.740, 0.758,
           0.724, 0.739, 0.736, 0.735, 0.760, 0.750, 0.759, 0.754, 0.744, 0.758, 0.765, 0.756,
           0.738, 0.742, 0.758, 0.757, 0.724, 0.757, 0.744, 0.738, 0.763, 0.756, 0.760, 0.768,
           0.761, 0.742, 0.734, 0.754, 0.758, 0.735, 0.740, 0.743, 0.737, 0.737, 0.725, 0.761,
           0.758, 0.756]
```

```python
# valores ordenados
ordenados = pd.Series(valores).sort_values()
```

```python
# dicionário de dados
parafuso = {
    "Comprimento Parafusos 3/4 polegada" : ["0,720-0,729", "0,730-0,739", "0,740-0,749", "0,750-0,759",
                                            "0,760-0,769"]
}
```

```python
# categorizar os dados por classes
bins = [0.720, 0.730, 0.740, 0.750, 0.760, 0.770]
cat = pd.cut(x = ordenados, bins = bins, labels = parafuso["Comprimento Parafusos 3/4 polegada"], include_lowest = True, right = False)
```

```python
# contagem de frequência por classes
freq = cat.value_counts().sort_index()
freq
```

```python
# distribuição de frequência
df = pd.DataFrame(parafuso)
df["Frequência"] = freq.tolist()
df["Frequência Relativa"] = ((df["Frequência"] / df["Frequência"].sum())*100).round(3)
df["Frequência Acumulada"] = df["Frequência"].cumsum()
df
```

```python
# fronteiras de classe
front1 = float(input("Qual a primeira fronteira de classe?"))
ampltd = float(input("Qual a amplitude de classes?"))
front2 = float(input("Qual a última fronteira de classe?"))

ff = []
while front1 <= front2:
    ff.append(round(front1, 4))
    front1+=ampltd
ff
```

```python
# plotar histograma
plt.title("Análise - Comprimento de Parafusos 3/4 de polegada")
plt.xlabel("Comprimento Parafusos 3/4 polegada")
plt.ylabel("Frequência")
plt.hist(x = valores, bins = ff, color = 'skyblue', edgecolor = 'black')
plt.xticks(ff)
plt.ylim(0, 18)
plt.show()
```

### O que o Histograma sugere sobre o comprimento de 3/4 de polegada, conforme escrito nos rótulos das embalagens desses parafusos?

Como podemos ver, a maior parte das medições resultaram em algo entre 0,7395 e 0,7595 (0,7395 + 0,7595 / 2 = 0,7495), ou seja, o comprimento da maior parte dos parafusos estão corretos.
