### Exercício 18. Coca Normal e Coca Dietética
#### Use a distribuição de frequência do Exercício 26, na seção 2-2, para contruir o histograma para os pesos de Coca normal e outro histograma para os pesos de Coca dietética.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# versões bibliotecas
print(f'Versão Pandas: {pd.__version__}')
print(f'Versão Matplotlib: {matplotlib.__version__}')
```

```python
# valores exercício

valores_diet = [0.7758, 0.7760, 0.7771, 0.7773, 0.7802, 0.7806, 0.7811, 0.7813, 0.7822, 0.7822,
                  0.7822, 0.7826, 0.7830, 0.7833, 0.7837, 0.7839, 0.7844, 0.7852, 0.7852, 0.7859,
                  0.7861, 0.7868, 0.7870, 0.7872, 0.7874, 0.7874, 0.7879, 0.7879, 0.7881, 0.7885,
                  0.7892, 0.7896, 0.7907, 0.7910, 0.7923, 0.7923]

valores_normal = [0.7901, 0.8044, 0.8062, 0.8073, 0.8079, 0.8110, 0.8126, 0.8128, 0.8143, 0.8150,
                0.8150, 0.8152, 0.8152, 0.8161, 0.8161, 0.8163, 0.8165, 0.8170, 0.8172, 0.8176,
                0.8181, 0.8189, 0.8192, 0.8192, 0.8194, 0.8194, 0.8207, 0.8211, 0.8229, 0.8244,
                0.8244, 0.8247, 0.8251, 0.8264, 0.8284, 0.8295]
```

```python
# valores ordenados

ordenados_normal = pd.Series(valores_normal).sort_values()
ordenados_diet = pd.Series(valores_diet).sort_values()
```

```python
# dicionários de classes

normal = {
    "Pesos Coca Normal (lb)" : ['0,7900 - 0,7949', '0,7950 - 0,7999', '0,8000 - 0,8049', '0,8050 - 0,8099', '0,8100 - 0,8149', '0,8150 - 0,8199', '0,8200 - 0,8249', '0,8250 - 0,8299']
}

diet = {
    "Pesos Coca Diet (lb)" : ['0,7750 - 0,7799', '0,7800 - 0,7849', '0,7850 - 0,7899', '0,7900 - 0,7949']
}
```

```python
# categorizar dados por classes
bins_normal = [0.7900, 0.7950, 0.8000, 0.8050, 0.8100, 0.8150, 0.8200, 0.8250, 0.8300]
cat_normal = pd.cut(x = ordenados_normal, bins = bins_normal, labels = normal["Pesos Coca Normal (lb)"], include_lowest = True, right = False)

bins_diet = [0.7750, 0.7800, 0.7850, 0.7900, 0.7950]
cat_diet = pd.cut(x = ordenados_diet, bins = bins_diet, labels = diet["Pesos Coca Diet (lb)"], include_lowest = True, right = False)
```

```python
# frequência dos dados por classes
freq_normal = cat_normal.value_counts().sort_index()
freq_diet = cat_diet.value_counts().sort_index()
```

```python
# distribuição de frequência completa - Coca Normal
df_normal = pd.DataFrame(normal)
df_normal['Frequência'] = freq_normal.tolist()
df_normal['Frequência Relativa'] = ((df_normal['Frequência'] / df_normal['Frequência'].sum())*100).round(2)
df_normal['Frequência Acumulada'] = df_normal['Frequência'].cumsum()
df_normal
```

```python
# distribuição de frequência completa - Coca Diet
df_diet = pd.DataFrame(diet)
df_diet['Frequência'] = freq_diet.tolist()
df_diet['Frequência Relativa'] = ((df_diet['Frequência'] / df_diet['Frequência'].sum())*100).round(2)
df_diet['Frequência Acumulada'] = df_diet['Frequência'].cumsum()
df_diet
```

```python
# fronteiras de classe - Coca Normal
front1 = float(input('Qual a primeira fronteira de classe?'))
ampltd = float(input('Qual a amplitude de classe?'))
front2 = float(input('Qual a última fronteira de classe?'))

ff = []
while front1 <= front2:
    ff.append(round(front1, 5))
    front1+=ampltd
ff
```

```python
# fronteiras de classe - Coca Diet
front1 = float(input('Qual a primeira fronteira de classe?'))
ampltd = float(input('Qual a amplitude de classe?'))
front2 = float(input('Qual a última fronteira de classe?'))

ff2 = []
while front1 <= front2:
    ff2.append(round(front1, 5))
    front1+=ampltd
ff2
```

```python
# plotando histogramas
# Cria uma figura e 2 eixos (subplots) em 1 linha e 2 colunas
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))

# --- Plotando o primeiro histograma (no eixo da esquerda) ---
axs[0].hist(x = valores_normal, bins= bins_normal, color='skyblue', edgecolor='black')
axs[0].set_title('Análise - Pesos Coca Normal')
axs[0].set_xlabel('Pesos Coca Normal')
axs[0].set_ylabel('Frequência')
axs[0].set_ylim([0, 20]) # Define o mesmo limite X para facilitar comparação
axs[0].set_xticks(ff)

# --- Plotando o segundo histograma (no eixo da direita) ---
axs[1].hist(x = valores_diet, bins= bins_diet, color='salmon', edgecolor='black')
axs[1].set_title('Análise - Pesos Coca Diet')
axs[1].set_xlabel('Pesos Coca Diet')
axs[1].set_ylabel('Frequência')
axs[1].set_ylim([0, 20]) # Define o mesmo limite X
axs[1].set_xticks(ff2)

plt.tight_layout() # Ajusta o layout automaticamente
plt.show()
```

### Compare os resultados e determine se parece haver uma diferença significante.

1. Observando o gráfico da esquerda, observamos que os pesos da Coca Normal geralmente estão em 0,81995 (0,8150 + 0,8249 / 2 = 0,81995).
2. Observando o gráfico da direita, observamos que os pesos da Coca Diet geralmente estão em 0,78495 (0,7800 + 0,7899 / 2 = 0,78495).

Os resultados fazem total sentido, já que a Coca Diet, diferente da Normal, não possui açucares. Logo ela tem de ser mais leve do que a Coca Normal.
