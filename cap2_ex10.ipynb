### Exercício 10. Radiação em Dentes de Crianças
#### Use a distribuição de frequência do Exercício 18, na seção 2-2, para construir um histograma

```python
# Importando Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Valores do exercício
valores = [155, 142, 149, 130, 151, 163, 151, 142, 156, 133, 138, 161, 128, 144, 172, 137, 151, 166, 147, 163, 145, 116, 136, 158, 114, 165, 169, 145, 150, 150, 150, 158, 151, 145, 152, 140, 170, 129, 188, 156]
```

```python
# dicionário de dados
dados = {
    'Estrôncio-90 (em milibecquerels)': ['110-119', '120-129', '130-139', '140-149', '150-159', '160-169', '170-179', '180-189']
}
```

```python
# criar um Series com os valores ordenados
freq = pd.Series(valores).sort_values()

# categoriza os dados em classes
bins = [110,120,130,140,150,160,170,180,190]
freq2 = pd.cut(x = freq, bins = bins, labels = dados['Estrôncio-90 (em milibecquerels)'], include_lowest=True, right=False)

# faz a contagem dos valores por classes (o resultado precisa ter o mesmo tamanho de classes que o dicionário)
freq_real = freq2.value_counts().sort_index()
```

bins=[110,120,130,140,150,160,170,180,190]
→ cria faixas de 110–119, 120–129, ..., 180–189 (pois right=False exclui o limite direito).

labels=classes
→ usa os rótulos do dicionário.

include_lowest=True
→ faz com que o limite inferior seja inclusivo.

right=False
→ faz com que o limite superior seja exclusivo.

```python
# conferindo os resultados
freq_real
```

```python
# gerando a distribuição de frequência dos dados
df = pd.DataFrame(dados)
df['Frequência'] = freq_real.tolist()
df['Frequência Relativa'] = ((df['Frequência'] / df['Frequência'].sum())*100).round(2)
df['Frequência Acumulada'] = df['Frequência'].cumsum()

df
```

### Plotando o Histograma

```python
# gerar fronteiras de classe
front1 = float(input("Digite a primeira fronteira de classe: "))
ampltd = float(input("Qual a amplitude de classe? "))
front2 = float(input("Digite a última fronteira de classe: "))

ff = []
while front1 <= front2:
    ff.append(front1)
    front1+=ampltd
    
ff
```

```python
# gerando histograma
plt.title('Análise de Radiação em Dentes de Crianças')
plt.xlabel('Estrôncio-90 (em milibecquerels)')
plt.ylabel('Frequência')
plt.hist(valores, bins = bins, rwidth = 0.9, color = 'green', edgecolor = 'black')
plt.xticks(ff)
plt.ylim(0, 15)
plt.show()
```
