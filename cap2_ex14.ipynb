### Exercício 14. Medidas de Voltagem Gerador
#### Use a distribuição de frequência do Exercício 22, na seção 2-2, para contruir o histograma.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print(f"Versão Pandas: {pd.__version__}")
print(f"Versão Matplotlib: {matplotlib.__version__}")
```

```python
# valores do exercício
valores = [124.8, 124.3, 125.2, 124.5, 125.1, 124.8, 125.1, 125.0, 124.8, 124.7, 124.5, 125.2, 124.4, 124.7,
           124.9, 124.5, 124.8, 124.8, 124.5, 124.6, 125.0, 124.7, 124.9, 124.9, 124.7, 124.2, 124.7, 124.8,
           124.4, 124.6, 124.4, 124.0, 124.7, 124.4, 124.6, 124.6, 124.8, 124.3, 124.0]
```

```python
# valores ordenados
ordenados = pd.Series(valores).sort_values()
ordenados
```

```python
# dicionário de dados
volt = {
    "Medidas de Voltagem Gerador" : ['123,90-124,09', '124,10-124,29', '124,30-124,49', '124,50-124,69', '124,70-124,89', '124,90-125,09', '125,10-125,29']
}
```

```python
# categoriza os dados em classes
bins = [123.9, 124.1, 124.3, 124.5, 124.7, 124.9, 125.1, 125.3]
freq = pd.cut(x = valores, bins = bins, labels = volt["Medidas de Voltagem Gerador"], include_lowest = True, right = False)

# contagem de dados por classes
freq2 = freq.value_counts().sort_index()
freq2
```

```python
# gerar distribuição de frequência
df = pd.DataFrame(volt)
df["Frequência"] = freq2.tolist()
df["Frequência Relativa"] = ((df["Frequência"] / df["Frequência"].sum())*100).round(2)
df["Frequência Acumulada"] = df["Frequência"].cumsum()
df
```

```python
# fronteiras de classe
front1 = float(input("Qual a primeira fronteira de classe?"))
ampltd = float(input("Qual a amplitude de classe?"))
front2 = float(input("Qual a última fronteira de classe?"))

ff = []
while front1 <= front2:
    ff.append(round(front1, 3))
    front1+=ampltd
ff
```

```python
# plotar Histograma
plt.title("Ánalise - Voltagem Gerador")
plt.xlabel("Medidas de Voltagem Gerador")
plt.ylabel("Frequência")
plt.hist(x = valores, bins = bins, color = 'orange', edgecolor = 'black', rwidth = 0.9)
plt.xticks(ff)
plt.ylim(0,14)
plt.show()
```

### Usando uma interpretação bem relaxada dos critérios relevantes, o resultado parece ser uma distribuição normal? Compare esse Histograma com o do exercício anterior.

Sim. No exercício anterior notamos que a frequência dos dados para cada classe são bens parecidos, gerando barras de mesmo tamanho. Nesse histograma, ja conseguimos notar uma distribuição normal nos dados, a frequência cresce, tem um pico e decai com uma força parecida com a da subida.
