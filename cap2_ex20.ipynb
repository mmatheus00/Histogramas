### Exercício 20. Pesos de Moedas de Quarto de Dólar
#### Use a distribuição de frequência do Exercício 28, na seção 2-2, para contruir o histograma.

```python
# importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# versões bibliotecas
print(f'Versão Pandas: {pd.__version__}')
print(f'Versão Matplotlib: {matplotlib.__version__}')
```

```python
# valores do exercício
valores = [5.5361, 5.5454, 5.5491, 5.5591, 5.5609, 5.5636, 5.5709, 5.5804, 5.5864, 5.5928, 5.5941,
           5.5999, 5.6010, 5.6022, 5.6063, 5.6114, 5.6157, 5.6160, 5.6234, 5.6274, 5.6449, 5.6485,
           5.6486, 5.6555, 5.6646, 5.6661, 5.6668, 5.6694, 5.6703, 5.6841, 5.6848, 5.6872, 5.7027,
           5.7050, 5.7198, 5.7239, 5.7247, 5.7344, 5.7495, 5.7790]
```

```python
# valores ordenados
ordenados = pd.Series(valores).sort_values()
```

```python
# dicionário de classes
classes = {
    "Pesos de Moedas Pós-1964" : ['5.5000 - 5.5499', '5.5500 - 5.5999', '5.6000 - 5.6499', '5.6500 - 5.6999', '5.7000 - 5.7499', '5.7500 - 5.7999']
}
```

```python
# segmentar os dados por classes
bins = [5.5000, 5.5500, 5.6000, 5.6500, 5.7000, 5.7500, 5.8000]
cat = pd.cut(x = valores, bins = bins, labels = classes["Pesos de Moedas Pós-1964"], include_lowest = True, right = False)
```

```python
# gerar frequências por classe
freq = cat.value_counts().sort_index()
```

```python
# gerar distribuição de frequência completa
df = pd.DataFrame(classes)
df["Frequência"] = freq.tolist()
df["Frequência Relativa"] = ((df["Frequência"] / df["Frequência"].sum())*100).round(2)
df["Frequência Acumulada"] = df["Frequência"].cumsum()
df
```

```python
# gerar pontos médios de classe
ponto1 = bins[0]
ampltd = (bins[1] - bins[0])
ponto2 = bins[1] - 0.0001
classe = len(df)

pm = ((ponto1 + ponto2) / 2)

pf = []
n = 1
while n <= classe:
    pf.append(round(pm, 5))
    pm+=ampltd
    n+=1
pf
```

```python
# plotar histograma
plt.title("Análise - Pesos de Moedas de Quarto de Dólar Pós-1964")
plt.xlabel("Pesos de Moedas Pós-1964")
plt.ylabel("Frequência")
plt.hist(x = valores, bins = bins, color = 'salmon', edgecolor = 'black')
plt.ylim(0, 12)
plt.xticks(pf)
plt.show()
```

### Compare esse histograma com o do exercício anterior

Podemos observar que o peso das moedas de quarto de dólar (25 cents) diminuiram significativamente após 1964.
O fenômeno do porque isso acontece está descrito no conjunto de dados, onde é explicado que até 1964, as moedas tinham 90% de prata e 10% de cobre em sua composição. E após 1964, as moedas começaram a ser feitas de uma liga de cobre, um material muito mais leve.
