```python
# categorizar dados por classes
bins = [400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900]
cat = pd.cut(x = ordenados, bins = bins, labels = classes["Escores FICO"], include_lowest = True, right = False)
```

```python
# frequência de dados por classes
freq = cat.value_counts().sort_index()
freq
```
