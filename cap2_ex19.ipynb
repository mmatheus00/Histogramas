### Exercício 19. Pesos de Moedas de Quarto de Dólar
#### Use a distribuição de frequência do Exercício 27, na seção 2-2, para contruir o histograma.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# versões bibliotecas
print(f'Versão Pandas: {pd.__version__}')
print(f'Versão Matplotlib: {matplotlib.__version__}')
```

```python
# valores do exercício
valores = [6.0002, 6.0257, 6.0760, 6.0775, 6.0829, 6.1073, 6.1095, 6.1181, 6.1275, 6.1309, 6.1352,
           6.1423, 6.1426, 6.1442, 6.1487, 6.1501, 6.1661, 6.1719, 6.1787, 6.1940, 6.1947, 6.1949,
           6.1970, 6.2130, 6.2151, 6.2371, 6.2412, 6.2441, 6.2465, 6.2647, 6.2674, 6.2718, 6.2771,
           6.2821, 6.2866, 6.2908, 6.3172, 6.3278, 6.3415, 6.3669]
```

```python
# valores ordenados
ordenados = pd.Series(valores).sort_values()
```

```python
# dicionário de Classes
classes = {
    "Pesos de Moedas Pré-1964" : ['6.0000 - 6.0499', '6.0500 - 6.0999', '6.1000 - 6.1499', '6.1500 - 6.1999', '6.2000 - 6.2499',
                                '6.2500 - 6.2999', '6.3000 - 6.3499', '6.3500 - 6.3999']
}
```

```python
# segmentar os dados dentro das classes
bins = [6.0000, 6.0500, 6.1000, 6.1500, 6.2000, 6.2500, 6.3000, 6.3500, 6.4000]
categorias = pd.cut(x = valores, bins = bins, labels = classes["Pesos de Moedas Pré-1964"], include_lowest = True, right = False)
```

```python
# gerar a frequência dos dados
freq = categorias.value_counts().sort_index()
```

```python
# gerar distribuição de frequência completa
df = pd.DataFrame(classes)
df['Frequência'] = freq.tolist()
df['Frequência Relativa'] = ((df['Frequência'] / df['Frequência'].sum())*100).round(2)
df['Frequência Acumulada'] = df['Frequência'].cumsum()
df
```

```python
# gerar fronteiras de classe
front1 = float(input("Qual a primeira fronteira de classe?"))
ampltd = float(input("Qual a amplitude de classe?"))
front2 = float(input("Qual a última fronteira de classe?"))

ff = []
while front1 <= front2:
    ff.append(round(front1, 5))
    front1+=ampltd
ff
```

```python
# plotar histograma
plt.title("Análise - Pesos de Moedas de Quarto de Dólar Pré-1964")
plt.xlabel("Pesos de Moedas Pré-1964")
plt.ylabel("Frequência")
plt.hist(x = valores, bins = bins, color = 'salmon', edgecolor = 'black')
plt.xticks(ff)
plt.show()
```
