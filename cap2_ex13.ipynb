### Exercício 13. Medidas de Voltagem Domiciliar
#### Use a distribuição de frequência do Exercício 21, na seção 2-2, para contruir o histograma.

```python
# importando bibliotecas
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
print(f"Versão do Pandas: {pd.__version__}")
print(f"Versão do Matplotlib: {matplotlib.__version__}")
```

```python
# valores do exercício
valores = [123.8, 123.9, 123.9, 123.3, 123.4, 123.3, 123.6, 123.5, 123.5, 123.5, 123.7, 123.6, 123.7, 123.9, 124.0, 124.2, 123.9, 123.8, 123.8, 124.0, 123.9, 123.6, 123.5, 123.4, 123.4, 123.4, 123.4, 123.3, 123.5, 123.5, 123.6, 123.8, 123.9, 123.9, 123.8, 123.9, 123.7, 123.8, 123.8, 123.3]
```

```python
# dicionário de dados
volt = {
    "Medidas de Voltagem Domiciliar" : ['123.3-123.49', '123.5-123.69', '123.7-123.89', '123.9-124.09', '124.1-124.29']
}
```

```python
# Series com valores ordenados
ordenados = pd.Series(valores).sort_values()
```

```python
# categorizar dados por classes
bins = [123.3, 123.5, 123.7, 123.9, 124.1, 124.3]
freq = pd.cut(x = ordenados, bins = bins, labels = volt["Medidas de Voltagem Domiciliar"], include_lowest = True, right = False)
```

```python
# contagem de dados por classes
freq2 = freq.value_counts().sort_index()
```

```python
### gerar distribuição de frequÊncia
df = pd.DataFrame(volt)
df['Frequência'] = freq2.tolist()
df['Frequência Relativa'] = ((df['Frequência'] / df['Frequência'].sum())*100).round(2)
df['Frequência Acumulada'] = df['Frequência'].cumsum()
df
```

```python
# gerar fronteiras de classe
front1 = float(input("Qual a primeira fronteira de classe?"))
ampltd = float(input("Qual a amplitude de classe?"))
front2 = float(input("Qual a última fronteira de classe?"))

ff = []
while front1 <= front2:
    ff.append(round(front1, 3))
    front1+=ampltd

ff
```

```python
# gerar histograma
plt.title('Análise - Voltagem Domiciliar')
plt.xlabel('Medidas de Voltagens')
plt.ylabel('Frequência')
plt.hist(x = valores, bins = bins, rwidth = 0.9, color = 'orange', edgecolor = 'black')
plt.xticks(ff)
plt.ylim(0, 11)
plt.show()
```

### O resultado parece ser uma distribuição normal? Por que sim ou por que não?

Não. A medição dos dados estão bem parecidas, não havendo grandes diferenças entre as classes e gerando barras de tamanhos iguais.
